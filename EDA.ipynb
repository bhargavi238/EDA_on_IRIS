### Sepal Length vs. Petal Length:
- There is a clear positive correlation between sepal length and petal length.
- Iris flowers with longer sepals also tend to have longer petals, while flowers with shorter sepals have shorter petals. 
### Sepal Width vs. Petal Width:
- There is a relatively weak correlation between sepal width and petal width.

# Insights from heat map
Sepal Length and Petal Length:
  There is a strong positive correlation between sepal length and petal length .
  As the sepal length increases, the petal length tends to increase as well.
  This indicates that longer sepals are associated with longer petals.
Sepal Length and Petal Width:
  There is a moderate positive correlation between sepal length and petal width .
  As the sepal length increases, the petal width tends to increase to some extent.
Petal Length and Petal Width:
  There is a strong positive correlation between petal length and petal width (high positive correlation coefficient).
  As the petal length increases, the petal width tends to increase as well.
Sepal Width and Petal Width:
  There is a weak negative correlation between sepal width and petal width (weak negative correlation coefficient).
  As the sepal width increases, the petal width tends to decrease slightly.
  This indicates that wider sepals are associated with narrower petals, although the correlation is not very strong.

```python
from sklearn.linear_model import LinearRegression

# Separate the Iris dataset by species
setosa = df[df['class'] == 'Iris-setosa']
versicolor = df[df['class'] == 'Iris-versicolor']
virginica = df[df['class'] == 'Iris-virginica']


sns.scatterplot(data=df, x='sepallength', y='sepalwidth', hue='class')

# Fit and plot regression lines for each species
reg_setosa = LinearRegression().fit(setosa[['sepallength']], setosa['sepalwidth'])
sns.lineplot(x=setosa['sepallength'], y=reg_setosa.predict(setosa[['sepallength']]), color='blue', label='Setosa')

reg_versicolor = LinearRegression().fit(versicolor[['sepallength']], versicolor['sepalwidth'])
sns.lineplot(x=versicolor['sepallength'], y=reg_versicolor.predict(versicolor[['sepallength']]), color='orange', label='Versicolor')

reg_virginica = LinearRegression().fit(virginica[['sepallength']], virginica['sepalwidth'])
sns.lineplot(x=virginica['sepallength'], y=reg_virginica.predict(virginica[['sepallength']]), color='green', label='Virginica')

# Set labels and title
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Multiple Regression Lines for Sepal Length vs Sepal Width')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
sns.scatterplot(data=df, x='petallength', y='petalwidth', hue='class')

# Fit and plot regression lines for each species
reg_setosa = LinearRegression().fit(setosa[['petallength']], setosa['petalwidth'])
sns.lineplot(x=setosa['petallength'], y=reg_setosa.predict(setosa[['petallength']]), color='blue', label='Setosa')

reg_versicolor = LinearRegression().fit(versicolor[['petallength']], versicolor['petalwidth'])
sns.lineplot(x=versicolor['petallength'], y=reg_versicolor.predict(versicolor[['petallength']]), color='orange', label='Versicolor')

reg_virginica = LinearRegression().fit(virginica[['petallength']], virginica['petalwidth'])
sns.lineplot(x=virginica['petallength'], y=reg_virginica.predict(virginica[['petallength']]), color='green', label='Virginica')

# Set labels and title
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Multiple Regression Lines for Sepal Length vs Sepal Width')

# Add legend
plt.legend()

# Show the plot
plt.show()
```
